## Ejercicio 2: Top 10 arrival airports in the world in 2013 (using the bookings file)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sn
import bz2
import csv
```

```python
bookings_sample=pd.read_csv ("bookings.csv.bz2", nrows=20000, sep="^")
```

```python
pd.options.display.max_columns=None
bookings_sample.head ()
```

```python
bookings_sample.columns
```

```python
bookings_sample_red= bookings_sample[['arr_port', 'pax', 'year']]
```

```python
bookings_sample_red['year']==2013
```

```python
bookings_sample_red_year=bookings_sample_red[bookings_sample_red['year']==2013]
```

```python
bookings_sample_red_year.shape
```

```python
bookings_sample_red_year
```

```python
bookings_sample_red_year.groupby('arr_port').sum()[['pax']].sort_values('pax', ascending=False).head(10)
```

```python
bookings_sample_red_year.groupby('arr_port').agg (suma= ('pax','sum')).sort_values ('suma', ascending=False).head(10)
```
